```python
from vllm import LLM, SamplingParams
```

```python
llm = LLM(
    model="../models/Qwen2.5-1.5B-Instruct",
    dtype="bfloat16",
    gpu_memory_utilization=0.80,
    max_model_len=2048,    
)
```

```python
sampling_params = SamplingParams(
    temperature=0.7,
    top_p=0.95,
    max_tokens=256,
)
```

```python
messages = [
    [{"role": "user", "content": "解释一下什么是注意力机制"}],
]

outputs = llm.chat(messages, sampling_params)
```
